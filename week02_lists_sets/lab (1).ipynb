## Task 5 – Introduction to Sets

Create a list containing the following IP addresses:

192.168.1.1  
192.168.1.2  
192.168.1.1  
10.0.0.5  
10.0.0.5  

Convert this list into a **set**.

Print the set.

```python
list = ["192.168.1.1", "192.168.1.2" , "192.168.1.1", "10.0.0.5", "10.0.0.5"]
set = set(list)
print(set)
```

## Task 6 – Unique Values

Using the set created in the previous task:

- Print each IP address on a separate line

```python
print("\n".join(set))
print(len(set))
```

## Task 7 – User Input with Lists

Ask the user to enter **five IP addresses**.

Store them in a list.

Then:

- Convert the list into a set
- Print the unique IP addresses
- Print the total number of unique IP addresses

```python
input_list = []
for i in range(5):
    ip_address = input(f"Enter IP address {i + 1}: ")
    input_list.append(ip_address)

unique_ips = (input_list)
print("\n".join(unique_ips))
print(len(unique_ips))
```
